```python
import ast
def converter(convert):
    lis=[]

    for feature in ast.literal_eval(convert):
        lis.append(feature['name'])
    return lis   
    
def fetch3(convert):
    counter =0
    L =[]
    for feature in ast.literal_eval(convert):
        if counter !=3:
            L.append(feature['name'])
            counter+=1
        else:
            break
    return L     
def fetch_director(convert):
    l =[]
    for feature in ast.literal_eval(convert):
        if feature['job'] == 'Director':
            
            l.append(feature['name'])
            break
    return l
    
```

```python
movies['genres'] = movies['genres'].apply(converter)
movies['keywords'] = movies['keywords'].apply(converter)
movies['cast'] = movies['cast'].apply(fetch3)
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['release_date'] = [i.split('-')[0] for i in movies['release_date']]
movies['release_date'] = pd.to_numeric(movies['release_date'])
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres']+ movies['keywords'] + movies['overview']+ movies['cast']+movies['crew']
```

```python

movies = movies.drop(columns=['genres','keywords','overview','cast','crew'])
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))

```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
Ps = PorterStemmer()
```

```python
def stem(obj):
    word=[]
    for i in obj.split():
        word.append(Ps.stem(i))
    return " ".join(word)  
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine = cosine_similarity(vector)
```
